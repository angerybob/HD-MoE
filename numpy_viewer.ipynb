```python
import numpy as np

# 定义文件路径，需根据实际情况修改
file_path = '/data/home/haochenhuang/deployment/results/reasoning_ds_10.0_TFLOPS_25.0_GBPS_for_8*8_mesh_128_batches/arrays_10.0_TFLOPS_25.0_GBPS_in_layer_0.npz'

try:
    np.set_printoptions(threshold=np.inf)
    # 加载 .npz 文件
    with np.load(file_path) as data:
        # 查看文件中包含的所有数组的名称
        array_names = data.files
        print("文件中包含的数组名称：")
        for name in array_names:
            print(name)

        # 查看每个数组的内容和形状
        for name in array_names:
            array = data[name]
            np.set_printoptions(threshold=np.inf)
            print(f"\n数组名称: {name}")
            print(f"数组形状: {array.shape}")
            print(f"数组内容: {array}")
except FileNotFoundError:
    print(f"错误：未找到文件 {file_path}")
except Exception as e:
    print(f"发生未知错误: {e}")
```
